Uses an integer as a bit vector to track seen lowercase letters (aâ€“z) by mapping each character to a bit position.
If a bit is already set, it returns False; otherwise it sets the bit and finally returns True (O(n) time, O(1) space).

```python
def isUnique(string: str):
    checker = 0   # bit vector

    for ch in string:
        if ch < 'a' or ch > 'z':
            continue   # skip non-lowercase chars (or handle separately)

        bit_index = ord(ch) - ord('a')

        # check if bit is already set
        if (checker & (1 << bit_index)) > 0:
            return False

        # set the bit
        checker |= (1 << bit_index)

    return True

print(isUnique("abc"))
print(isUnique("aba"))
```

```python
def isUnique(string: str):
  size = len(string)
  for i in range(size):
      if string[i] in string[i+1:]:
        return False
  return True

print(isUnique("44"))
print(isUnique("117"))
print(isUnique("132"))
print(isUnique("abc"))
print(isUnique("aba"))
print(isUnique("1234567890 aabba"))
```
